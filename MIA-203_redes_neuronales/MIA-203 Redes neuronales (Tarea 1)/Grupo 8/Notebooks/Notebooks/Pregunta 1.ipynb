```python
import numpy as np
```

```python
def tanh(x):
    """Función de activación tanh."""
    return np.tanh(x)

def tanh_derivative(x):
    """Derivada de la función tanh."""
    return 1 - np.tanh(x) ** 2
```

```python
X = np.array([
    [2.5, 3.5, -0.5],
    [4.0, -1.0, 0.5],
    [0.5, 1.5, 1.0],
    [3.0, 2.0, -1.5]
])

y = np.array([
    [1.0],
    [-1.0],
    [-1.0],
    [1.0]
])
print("x:", X)
print("y:", y)
```

```python
W1 = np.array([
    [-0.34108091502837, 0.4875444774924518, -0.5786687891714137],
    [-0.7406084700295616, -0.5419469493214886, -0.7157868250555495],
    [-0.4315505068114813, -0.3888294993216346, -0.7260616380268912],
    [0.8544519713956533, -0.24436625968891335, -0.29745436944045234]
])
b1 = np.array([
    [0.42066411352123567],
    [-0.06842417572242421],
    [0.3917250492343052],
    [0.07861738679252617]
])

W2 = np.array([
    [0.8240668394757076, -0.7925535106179675, 0.856718957818575, 0.38940887530348656],
    [-0.8401491947426882, 0.6005090770290116, -0.710761829952594, -0.5376371931922113],
    [-0.9617475006414558, -0.1484109682541901, 0.315131084013728, 0.843811249537862],
    [0.7589476929906058, -0.6751635852663345, -0.05422983414165872, 0.43000805343850246]
])
b2 = np.array([
    [0.2807627460701372],
    [0.6705844066805486],
    [0.2750476238785533],
    [0.20918398742213418]
])

W3 = np.array([
    [-0.3547048938594579, -0.3011280508308256, 0.5691627652457749, -0.9665772078201607]
])
b3 = np.array([[0.9979425023631463]])

print("W1:", W1)
print("W2:", W2)
print("W3:", W3)
print("b1:", b1)
print("b2:", b2)
print("b3:", b3)
```

```python
# Capa Oculta 1
z1 = W1.dot(X.T) + b1
a1 = tanh(z1)
print("z1:", z1)
print("a1:", a1)
```

```python
# Capa Oculta 2
z2 = W2.dot(a1) + b2
a2 = tanh(z2)
print("z2:", z2)
print("a2:", a2)
```

```python
# Capa de Salida
z3 = W3.dot(a2) + b3
a3 = tanh(z3)
print("z3:", z3)
print("a3:", a3)
```

```python
error = a3 - y.T
delta3 = 2 * error * tanh_derivative(z3)
print("error:", error)
print("delta3:", delta3)
```

```python
loss = np.sum((y.T - a3)**2)
print(loss)
```

```python
delta2 = W3.T.dot(delta3) * tanh_derivative(z2)
print("delta2:", delta2)
```

```python
delta1 = W2.T.dot(delta2) * tanh_derivative(z1)
print("delta1:", delta1)
```

```python
dW3 = delta3.dot(a2.T)
db3 = np.sum(delta3, axis=1, keepdims=True)

dW2 = delta2.dot(a1.T)
db2 = np.sum(delta2, axis=1, keepdims=True)

dW1 = delta1.dot(X)
db1 = np.sum(delta1, axis=1, keepdims=True)
```

```python
learning_rate = 0.1
W1 -= learning_rate * dW1
b1 -= learning_rate * db1
W2 -= learning_rate * dW2
b2 -= learning_rate * db2
W3 -= learning_rate * dW3
b3 -= learning_rate * db3
```

```python
print("\nUpdated Weights and Biases:")
print(f"W1: {W1}")
print(f"W2: {W2}")
print(f"W3: {W3}")
print(f"b1: {b1}")
print(f"b2: {b2}")
print(f"b3: {b3}")
```

```python
print("\nBackward Propagation Results:")
print(f"Gradientes de los pesos y biases:")
print(f"dW1: {dW1}")
print(f"db1: {db1}")
print(f"dW2: {dW2}")
print(f"db2: {db2}")
print(f"dW3: {dW3}")
print(f"db3: {db3}")
```
