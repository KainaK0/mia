```python
from ultralytics import YOLO
import os

# Cargar el modelo YOLOv11 y entrenar
model = YOLO('yolo11n.pt')  # Puedes cambiar a yolo11s.pt, yolo11m.pt, etc., según la variante que quieras usar

# Iniciar el entrenamiento
model.train(data='data.yaml', epochs=100, imgsz=960, batch=64, workers=16, amp=True)

```

```python

last_dir = model.trainer.save_dir

# Mostrar la ubicación del mejor modelo
best_model_path = os.path.join(last_dir, 'weights', 'best.pt')
print(f"El modelo final se guardó en: {best_model_path}")
```

```python
import os
import cv2
from matplotlib import pyplot as plt

# Cargar el modelo YOLOv11 que has entrenado
model = YOLO(best_model_path)  # Ajusta esta ruta al modelo entrenado
#model = YOLO("/content/runs/detect/train/weights/best.pt")

# Directorio donde están las imágenes
img_folder = 'Pruebas'

# Procesar todas las imágenes en la carpeta
for img_name in os.listdir(img_folder):
    if img_name.endswith(('.png', '.jpg', '.jpeg')):  # Asegurarse de que solo se lean imágenes
        img_path = os.path.join(img_folder, img_name)

        # Realizar la detección en la imagen
        results = model(img_path)

        # Mostrar los resultados
        results[0].show()

        # También puedes guardar la imagen con las detecciones
        results[0].save(f'output_{img_name}')  # Guardar cada imagen procesada con prefijo 'output_'
```

```python
from ultralytics import YOLO
import os

# Cargar el modelo YOLOv11
model = YOLO(best_model_path)

# Directorio donde se encuentran los videos MP4
video_folder = 'Pruebas'

# Procesar todos los archivos MP4 en la carpeta
for video_name in os.listdir(video_folder):
    if video_name.endswith('.mp4'):  # Verificar que sea un archivo MP4
        video_path = os.path.join(video_folder, video_name)

        # Realizar la detección en el video y guardar el resultado
        results = model(video_path, save=True)

        # Guardar la ruta del video procesado
        output_video_path = os.path.join(results[0].save_dir, video_name)

        # Mostrar la ubicación del video procesado
        print(f"El video procesado se ha guardado en: {output_video_path}")
```
